```python
import os
import pandas as pd

from agents.scenario_agent import (
    list_supported_scenarios,
    get_scenario_config,
    build_target_component_frame,
    build_target_outcome,
    attach_target_outcome,
    filter_historical_rows_for_scenario,
)

PROJECT_ROOT = os.path.abspath("..")
df = pd.read_csv(os.path.join(PROJECT_ROOT, "data/generated/historical_outcomes_raw.csv"))

print("Supported scenarios:", list_supported_scenarios())
print()
print("Scenario config:")
print(get_scenario_config("Supply Chain Disruption"))
print()

components = build_target_component_frame(df, "Supply Chain Disruption")
print("Component frame head:")
print(components.head())
print()

target = build_target_outcome(df, "Supply Chain Disruption")
print("Target head:")
print(target.head())
print()

print("Target summary:")
print(target.describe())
print()

filtered = filter_historical_rows_for_scenario(df, "Supply Chain Disruption")
print("Filtered rows for scenario:", len(filtered))
print(filtered[["scenario_type"]].head())
print()

df_with_target = attach_target_outcome(df, "Supply Chain Disruption")
print(df_with_target[["scenario_type", "target_scenario_score"]].head())
```
