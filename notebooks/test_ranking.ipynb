```python
import os
import sys
import pandas as pd

PROJECT_ROOT = os.path.abspath("..")
if PROJECT_ROOT not in sys.path:
    sys.path.insert(0, PROJECT_ROOT)

from agents.ranking_agent import (
    rank_candidates_for_scenario,
    fit_ranking_agent,
    run_standard_intervention_analysis,
)

historical_df = pd.read_csv(os.path.join(PROJECT_ROOT, "data/generated/historical_outcomes_raw.csv"))
current_df = pd.read_csv(os.path.join(PROJECT_ROOT, "data/generated/current_applicants_raw.csv"))

for scenario in [
    "Supply Chain Disruption",
    "Digital Transformation",
    "Operational Continuity",
]:
    ranked_df, summary, artifacts = rank_candidates_for_scenario(
        historical_df=historical_df,
        current_df=current_df,
        scenario_name=scenario,
        auto_assign_mechanisms=False,
    )

    print("\n" + "=" * 80)
    print("SCENARIO:", scenario)
    print("SUMMARY:", summary)
    print("ARTIFACTS:", artifacts)
    print(ranked_df[[
        "rank",
        "candidate_id",
        "name",
        "predicted_scenario_outcome",
        "scenario_alignment_score",
        "competency_disruption_management",
        "competency_digital_transformation",
        "competency_operational_continuity",
        "competency_leadership",
        "competency_bmw_context_fit",
    ]].head(5))

fitted_agent = fit_ranking_agent(
    historical_df=historical_df,
    scenario_name="Supply Chain Disruption",
    auto_assign_mechanisms=False,
)

ranked_df, _, _ = rank_candidates_for_scenario(
    historical_df=historical_df,
    current_df=current_df,
    scenario_name="Supply Chain Disruption",
    auto_assign_mechanisms=False,
)

what_if_df = run_standard_intervention_analysis(
    fitted_agent=fitted_agent,
    ranked_df=ranked_df,
    top_n=3,
    uplift=10.0,
)

print("\nWHAT-IF")
print(what_if_df.head(12))
```
